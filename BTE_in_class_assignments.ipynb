```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter numerical operator: ")



if operator == "+":
    res = num1 + num2
elif operator == "-":
    res = num1 - num2
elif operator == "*":
    res = num1 * num2
elif operator == "/":
    res = num1/num2
elif operator == "//":
    res = num1//num2
elif operator == "**":
  res = num1**num2
else:
  print("Invalid operator")

print(f"Result: {res}")
```
