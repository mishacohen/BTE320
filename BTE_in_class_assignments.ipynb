```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter numerical operator: ")



if operator == "+":
    res = num1 + num2
elif operator == "-":
    res = num1 - num2
elif operator == "*":
    res = num1 * num2
elif operator == "/":
    res = num1/num2
elif operator == "//":
    res = num1//num2
elif operator == "**":
  res = num1**num2
else:
  print("Invalid operator")

print(f"Result: {res}")
```

```python
import random

play = input("Do you want to play (y/n): ")

while play == "y":
  p_move = input("Choose between rock, paper, scissors: ").lower()
  c_move = random.choice(["rock", "paper", "scissors"])

  print(f"Player move: {p_move}\nComputer move: {c_move}")

  if p_move == c_move:
    print("Draw")
  elif (
      (p_move == "rock" and c_move == "scissors")or
      (p_move == "scissors" and c_move == "paper") or
      (p_move == "paper" and c_move == "rock")
  ):
    print("Player wins!")
  else:
    print("Computer wins")

  play = input("Do you want to play again (y/n): ")
```

```python
names = {}

while True:
  name = input("Give a name: ")
  if name == "":
    break

  if name in names:
    names[name] += 1
  else:
    names[name] = 1

print("Dictionary:")
print(names)
```
